## Step 3 — Create Anonymized Dataset

To ensure privacy and safe sharing of the dataset, personally identifiable information (PII) is removed or replaced.

This includes:
- Removing email, phone, and address fields
- Replacing names with generic labels
- Creating a new donor ID for tracking

```python
anon = df.copy()

# Replace names with generic donor labels
if "First Name" in anon.columns:
    anon["First Name"] = "Donor"

if "Last Name" in anon.columns:
    anon["Last Name"] = anon.index.astype(str)

# Remove personal info columns if they exist
for col in ["Email", "email", "Phone", "phone", "Address", "address"]:
    if col in anon.columns:
        anon[col] = None

# Create new donor_id
anon["donor_id"] = "donor_" + anon.index.astype(str)

anon.head()
```

### Observations

The anonymized dataset was successfully created from the consolidated dataset.

- Personally identifiable information (PII), such as email, phone, and address fields, has been removed or replaced.
- Donor names were replaced with generic labels to ensure privacy.
- A new `donor_id` field was created using the dataset index to maintain a unique identifier for each record.
- The dataset structure remains consistent, preserving key analytical fields such as `date`, `donation_amount`, `source_system`, and `source_category`.

This transformation ensures that:
- The dataset can be safely shared and uploaded to GitHub
- Donor activity can still be tracked without exposing sensitive information
- The data remains suitable for donor-level analysis and aggregation

Next, I will save the anonymized dataset for future use and continue with donor-level analysis.

## Step 6 Donor-Level Summary

To better understand donor behavior, I created a donor-level summary dataset.
This dataset aggregates donation activity at the donor level and includes:

- Total amount donated
- Number of donations
- First donation date
- Last donation date

This allows analysis of repeat donors, top donors, and donation frequency.

```python
donor_summary = df.groupby('donor_id').agg(
    total_donated=('donation_amount', 'sum'),
    number_of_donations=('donation_amount', 'count'),
    first_donation=('date', 'min'),
    last_donation=('date', 'max')
).reset_index()

donor_summary.head()
```

### Observations

The donor-level summary dataset was successfully created.

- Each row represents a unique donor identified by `donor_id`
- Aggregated metrics include total donation amount, number of donations, and donation date range
- This transformation converts transactional data into a structured format for analysis

Key observations:
- Many donors currently show zero total donations and donation counts, indicating missing or incomplete donation data
- Donation dates are present for some records, but gaps exist due to missing values in the original dataset
- The dataset provides a foundation for identifying repeat donors and understanding donor engagement

This dataset enables:
- Identification of top donors
- Analysis of donation frequency
- Tracking donor activity over time

Next, I will use this dataset to explore donation patterns and generate insights.

```python
donor_summary.to_csv("../data/outputs/donor_level_summary.csv", index=False)
print("Donor summary saved")
```
